# Background Segmentation
This notebook test the background segmentation system.

```python
from pathlib import Path

import matplotlib.pyplot as plt

from histokit.io.slides import get_slide_cls_for_path
from histokit.segmentation.detectors import per_patch_canny_segmentation

# Load the slide
slide_path = Path("../data/icaird/cervical_mini/slides/IC-CX-00002-01.ome.tiff")
SlideCls = get_slide_cls_for_path(slide_path)

with SlideCls(slide_path) as slide:
    # Get the detector from the registry and run it
    detector = per_patch_canny_segmentation(patch_size=224, patch_level=1)
    mask = detector(slide)

    # Get a thumbnail matching the mask dimensions
    thumb = slide.get_thumbnail_for_size(mask.shape[1], mask.shape[0])

    # Show the thumbnail and segmentation mask side by side
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
    ax1.imshow(thumb)
    ax1.set_title("Slide thumbnail")
    ax1.axis("off")
    ax2.imshow(mask, cmap="gray")
    ax2.set_title("per_patch_canny_segmentation")
    ax2.axis("off")
    plt.tight_layout()
    plt.show()
```
