# Pipeline
This notebook will test the pipeline system but running one of the default pipelines.

```python
from histokit.dataset import Dataset
from histokit.pipelines.presets.basic import pipeline

cerv_mini_dataset = Dataset.from_index(
    "../data/icaird/cervical_mini/index.csv",
    "../data/icaird/cervical_mini/labels.json",
    slides_dir="slides",
    annotations_dir="annotations"
)

patchsets = pipeline.run(cerv_mini_dataset, patch_size=224, tissue_threshold=0.02)
for patchset in patchsets:
    print(patchset.describe())
```
